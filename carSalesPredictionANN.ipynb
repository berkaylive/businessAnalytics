```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('car_purchasing.csv', encoding='latin-1')
```

```python
data.head()
```

```python
data.drop(columns=['customer name', 'customer e-mail', 'country', 'gender'], inplace=True)
```

```python
data.head()
```

```python
x = data.drop(columns=['car purchase amount'], axis=1)
y = data['car purchase amount']
```

```python
scaler = MinMaxScaler()
```

```python
x = scaler.fit_transform(x)
y = scaler.fit_transform(y.values.reshape(-1,1))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
#Model Structure Deep Learning
```

```python
model = Sequential()
```

```python
model.add(Dense(10, activation='relu', input_dim=4))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
#Model Training
```

```python
history = model.fit(x_train, y_train, epochs=100, validation_split=0.2)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, y_pred)
```

```python
print('The R2 score is {}'.format(score))
```
