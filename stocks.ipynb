```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
sns.set_style('whitegrid')
plt.style.use("fivethirtyeight")
%matplotlib inline
```

```python
import yfinance as yf
```

```python
# The stocks we will be using
stocks = ['AAPL', 'GOOG', 'MSFT', 'AMZN', 'NVO', 'LLY', 'ME']
```

```python
# Setting the end date and starting date (1 year ago)
end = datetime.now()
start = datetime(end.year-5, end.month, end.day)
```

```python
#Getting the Stocks Information
dic = {}

for stock in stocks:
  dic[stock] = yf.download(stock, start, end)
```

```python
#Getting the APPLE stock info as dataframe
apple = dic['AAPL']
```

```python
apple.head()
```

```python
#Plotting the 1 Year period pf APPLE adj closing price
apple['Adj Close'].plot()
plt.show()
```

```python
#Plotting the Moving Averages of 90 Days along with the stock price
apple['Adj Close'].plot(label = 'Stock Price')
apple['Adj Close'].rolling(90).mean().plot(label = 'MA 30 days')
plt.legend()
```

```python
#Plotting the Percentage Change of APPLE
(apple['Adj Close'].pct_change() * 100).hist(bins=50)
plt.title('Apple No of % Change 5-Year Period')
plt.ylabel('Counts')
plt.xlabel('% Change')
plt.show()
```
